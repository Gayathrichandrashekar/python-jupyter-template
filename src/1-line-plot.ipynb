### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt

# Generate data
x = list(range(0, 11))  # x values from 0 to 10
y = [i**2 for i in x]   # y values are the square of x

# Create the line plot
plt.plot(x, y, color='blue', marker='o', linestyle='--', linewidth=2, label='y = x^2')

# Customize the plot
plt.title('Line Plot of x vs x^2')        # Title
plt.xlabel('x values')                    # X-axis label
plt.ylabel('y = x squared')               # Y-axis label
plt.grid(True)                            # Add grid
plt.legend()                              # Show legend

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plotting
plt.plot(x, y, label='Linear Growth', marker='o')
plt.title('Basic Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 12, 34]

plt.bar(categories, values, color='orange')
plt.title("Category Value Comparison")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
